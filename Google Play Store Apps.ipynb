```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data=pd.read_csv("C:/Users/priya/OneDrive/Desktop/Python Datasets and Projects/googleplaystore.csv")
data.head(10)
```

```python
data.info()
```

```python
data.isnull().sum()
```

Dropping the duplicates values from the dataset

```python
data["App"].duplicated().sum()
```

```python
data.drop_duplicates(["App"],inplace=True)
```

```python
data.info()
```

```python
data[data["Type"].isnull()]
```

```python
data[data["Content Rating"].isnull()]
```

```python
data.drop(10472,inplace=True)
```

Removing the nan values from dataset

```python
data["Rating"]=data["Rating"].replace(np.nan,"0")
data["Type"]=data["Type"].replace(np.nan,"Not Mentioned")
data["Current Ver"]=data["Current Ver"].replace(np.nan,"Not Mentioned")
data["Android Ver"]=data["Android Ver"].replace(np.nan,"Not Mentioned")
data.isnull().sum()
```

Make the datatype correct whose have object but have to be int

```python
data["Reviews"]=data["Reviews"].astype(int)
data["Rating"]=pd.to_numeric(data["Rating"],errors="coerce").astype(float)
data.info()
```

```python
data["Installs"] = data["Installs"].astype(str)
data["Installs"] = data["Installs"].str.replace("+", "", regex=False)
data["Installs"] = data["Installs"].str.replace(",", "", regex=False)
data["Installs"] = pd.to_numeric(data["Installs"], errors="coerce").astype(int)
data.info()
```

```python
data["Price"] = data["Price"].str.replace("$", "", regex=False)
data["Price"] = pd.to_numeric(data["Price"], errors="coerce").astype(int)
data.info()
```

```python
data["Category"].unique()
```

```python
a=data.groupby("Category")["App"].count().sort_values(ascending=False).head(10)
plt.bar(a.index,a.values)
plt.xticks(rotation=90)
plt.xlabel("Category of apps")
plt.ylabel("Count of apps")
plt.title("Top 10 Category of Apps")
plt.show()
```

It shows that there is most type of categories of apps are family

```python
a=data.groupby("Category")["Rating"].mean().sort_values(ascending=False).head(10)
plt.bar(a.index,a.values)
plt.xticks(rotation=90)
plt.xlabel("Category of apps")
plt.ylabel("Mean of rating")
plt.title("Top 10 highest average rating of categories of apps")
plt.show()
```

It shows that peoples not had difficulties and love the education type of category of app
